```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100,min_samples_split=100, random_state=1)

train=nifty50.iloc[:-100]
test=nifty50.iloc[-100:]


predictor=["Close","Volume","Open","High","Low"]
model.fit(train[predictor],train["Target"])




```

```python
from sklearn.metrics import precision_score

pres=model.predict(test[predictor])
```

```python
import pandas as pd 


pres=pd.Series(pres, index=test.index)
```

```python
precision_score(test["Target"],pres)
```

```python
def predict(train, test,predictor, model):
    model.fit(train[predictor],train["Target"])
    pres=model.predict(test[predictor])
    pres=pd.Series(pres, index=test.index , name="Predictions")
    combine=pd.concat([test["Target"],pres],axis=1)
    return combine





```

```python
def backtest(data , model , predictor , start=2500, step=250):
    all_predictions=[]
    

    for i in range (start, data.shape[0],step):
        train=data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions=predict(train, test ,predictor, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)    








```

```python
predictions=backtest(nifty50,model,predictor)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
horizons=[2,5,60,250,1000]
new_predictors=[]

for horizon in horizons:
    rolling_averages=nifty50.rolling(horizon).mean()
    

    ratio_column =f"Close_Ratio_{horizon}"
    nifty50[ratio_column]=nifty50["Close"]/rolling_averages["Close"]


    trend_column=f"Trend_{horizon}"
    nifty50[trend_column]=nifty50.shift(1).rolling(horizon).sum()["Target"]
    

    new_predictors+=[ratio_column , trend_column]









```

```python
nifty50=nifty50.dropna()
```

```python
nifty50
```

```python
model=RandomForestClassifier(n_estimators=200, min_samples_split=50,random_state=1)

```

```python
def predict(train, test,predictor, model):
    model.fit(train[predictor],train["Target"])
    pres=model.predict_proba(test[predictor])[:,1]
    pres[pres>= .6]=1
    pres[pres< .6 ]=0
    pres=pd.Series(pres, index=test.index , name="Predictions")
    combine=pd.concat([test["Target"],pres],axis=1)
    return combine



```

```python
predictions=backtest(nifty50, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```
